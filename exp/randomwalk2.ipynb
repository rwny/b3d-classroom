# Random Walk with Mirror/Pingpong Function

This notebook shows two random walks plotted together:
1. A standard random walk that can go in any direction
2. A 'pingpong' or 'mirror' version that bounces between -1 and 1

```python
import random
import matplotlib.pyplot as plt

n = 2000
x = []
for i in range(n):
   x.append(i)

yy = 0
y = [yy]        # Standard random walk
ymirror = [yy]  # Mirror/pingpong random walk
ystep = 0.1

for i in range(n-1):
   coin_flip = random.random()
   if coin_flip > 0.5:
      next_step = ystep
   else:
      next_step = -ystep

   # Standard random walk - can go beyond any limits
   next_y = y[-1] + next_step
   y.append(next_y)
   
   # Mirror/pingpong random walk - bounces between -1 and 1
   next_mirror = ymirror[-1] + next_step
   
   # Check if the value would go beyond boundaries
   if next_mirror > 1:
      # Bounce off the upper boundary (1)
      next_mirror = (2 - next_mirror) + ystep  # Mirror effect at y=1
   elif next_mirror < -1:
      # Bounce off the lower boundary (-1)
      next_mirror = (-2 - next_mirror) - ystep  # Mirror effect at y=-1
   
   ymirror.append(next_mirror)

# Create a combined plot in a single figure
plt.figure(figsize=(12, 6))

# Add highlighted background for region between -1 and 1
plt.axhspan(-1, 1, facecolor='lightyellow', alpha=0.3, label='Boundary Region')

plt.plot(x, y, 'b-', color='gray', alpha=0.7, label='Standard Random Walk')
plt.plot(x, ymirror, 'r-', alpha=0.7, label='Mirror/Pingpong Random Walk')
plt.hlines([-1, 1], 0, n-1, colors='darkred', linestyles='dashed')
plt.title('Comparison of Random Walk Types')
plt.xlabel('Step')
plt.ylabel('Position')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

## How the Mirror/Pingpong Function Works

When the random walk reaches a boundary (either -1 or 1), it "bounces" off that boundary like a ball bouncing off a wall. This is calculated as follows:

- If the value exceeds 1: `new_value = 2 - original_value`
- If the value goes below -1: `new_value = -2 - original_value`

This creates a reflection effect that keeps the value always between -1 and 1.

In the plot above, you can see how the red line (mirror/pingpong) stays contained within the highlighted boundary region while the blue line (standard) can go anywhere.

```python
import random
import matplotlib.pyplot as plt

n = 2050
x = []
for i in range(n):
    x.append(i)

y = [0]         # Standard random walk
ymirror = [0]   # Mirror/pingpong random walk
ystep = 0.1

# กำหนดตัวแปรขอบเขตใหม่
upper_bound = 4
lower_bound = 0

# กำหนดตัวแปรสำหรับค่าคงที่ในการสะท้อน
reflection_constant = 2.0

for i in range(n-1):
    coin_flip = random.random()
    if coin_flip > 0.5:
        next_step = ystep
    else:
        next_step = -ystep

    # Standard random walk - can go beyond any limits
    next_y = y[-1] + next_step
    y.append(next_y)

    # Mirror/pingpong random walk - bounces between lower_bound and upper_bound
    next_mirror = ymirror[-1] + next_step

    # Check if the value would go beyond boundaries
    if next_mirror > upper_bound:
        # Bounce off the upper boundary
        next_mirror = reflection_constant * upper_bound - next_mirror  # Mirror effect at upper_bound
    elif next_mirror < lower_bound:
        # Bounce off the lower boundary
        next_mirror = reflection_constant * lower_bound - next_mirror  # Mirror effect at lower_bound

    ymirror.append(next_mirror)

# Create a combined plot in a single figure
plt.figure(figsize=(12, 6))

# Add highlighted background for region between lower_bound and upper_bound
plt.axhspan(lower_bound, upper_bound, facecolor='green', alpha=0.3, label='Boundary Region')

plt.plot(x, y, 'b-', alpha=0.7, label='Standard Random Walk')
plt.plot(x, ymirror, 'r-', alpha=0.7, label='Mirror/Pingpong Random Walk')
plt.hlines([lower_bound, upper_bound], 0, n-1, colors='darkred', linestyles='dashed')
plt.title('Comparison of Random Walk Types')
plt.xlabel('Step')
plt.ylabel('Position')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import random
import matplotlib.pyplot as plt

n = 500
x = []
for i in range(n):
    x.append(i)

y_standard = [0]     # Standard random walk
y_mirror = [0]       # Mirror random walk
y_pinned = [0]       # Pinned random walk
ystep = 0.1

# กำหนดตัวแปรขอบเขต
upper_bound = 2.0
lower_bound = -1.5
reflection_constant = 2.0

for i in range(n-1):
    coin_flip = random.random()
    if coin_flip > 0.5:
        step = ystep
    else:
        step = -ystep

    # 1. Standard Random Walk
    next_standard = y_standard[-1] + step
    y_standard.append(next_standard)

    # 2. Mirror Random Walk
    next_mirror = y_mirror[-1] + step
    if next_mirror > upper_bound:
        next_mirror = reflection_constant * upper_bound - next_mirror
    elif next_mirror < lower_bound:
        next_mirror = reflection_constant * lower_bound - next_mirror
    y_mirror.append(next_mirror)

    # 3. Pinned Random Walk
    next_pinned = y_pinned[-1] + step
    if next_pinned > upper_bound:
        next_pinned = upper_bound
    elif next_pinned < lower_bound:
        next_pinned = lower_bound
    y_pinned.append(next_pinned)

# Create a combined plot in a single figure
plt.figure(figsize=(12, 4))

# Add highlighted background for region between lower_bound and upper_bound
plt.axhspan(lower_bound, upper_bound, facecolor='lightyellow', alpha=0.3, label='Boundary Region')

plt.plot(x, y_standard, 'b-', alpha=0.7, label='Standard Random Walk')
plt.plot(x, y_mirror, 'r-', alpha=0.7, label='Mirror Random Walk')
plt.plot(x, y_pinned, 'g-', alpha=0.7, label='Pinned Random Walk')
plt.hlines([lower_bound, upper_bound], 0, n-1, colors='darkred', linestyles='dashed')

plt.title('Comparison of Random Walk Types')
plt.xlabel('Step')
plt.ylabel('Position')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
