# Multiple Random Walks

This notebook shows multiple random walks plotted together for comparison.

```python
import numpy as np
import random
import matplotlib.pyplot as plt

n = 2500
# Create a list of x values from 0 to n-1
x = np.arange(0, n)

# Create different random walks
y1 = [0.0]  # Standard random walk with step=0.05
y2 = [0.0]  # Random walk with step=0.02
y3 = [0.0]  # Random walk with step=0.01

# Generate multiple random walks with different step sizes
for i in range(n-1):  
    # Random walk 1 - step size 0.05
    coin_flip1 = random.random()
    next_step1 = 0.05 if coin_flip1 > 0.5 else -0.05
    next_y1 = y1[-1] + next_step1
    y1.append(next_y1)
    
    # Random walk 2 - step size 0.02
    coin_flip2 = random.random()
    next_step2 = 0.02 if coin_flip2 > 0.5 else -0.02
    next_y2 = y2[-1] + next_step2
    y2.append(next_y2)
    
    # Random walk 3 - step size 0.01
    coin_flip3 = random.random()
    next_step3 = 0.01 if coin_flip3 > 0.5 else -0.01
    next_y3 = y3[-1] + next_step3
    y3.append(next_y3)

# Create a nice big plot with all random walks
plt.figure(figsize=(12, 6))

# Add highlighted background for region between -1 and 1
plt.axhspan(-1, 1, facecolor='lightyellow', alpha=0.3, label='Reference Region (-1 to 1)')

plt.plot(x, y1, 'b-', alpha=0.7, label='Step Size = 0.05')
plt.plot(x, y2, 'r-', alpha=0.7, label='Step Size = 0.02')
plt.plot(x, y3, 'g-', alpha=0.7, label='Step Size = 0.01')
plt.hlines([-1, 1], 0, n-1, colors='darkgray', linestyles='dashed')
plt.title('Comparison of Random Walks with Different Step Sizes')
plt.xlabel('Step Number')
plt.ylabel('Position')
plt.grid(True)
plt.legend()
plt.show()
```

## 1D Random Walk

A random walk is a mathematical object that describes a path consisting of a succession of random steps. In this example, at each step we randomly decide whether to go up or down by a fixed amount.

```python
# 1D Random Walk
steps = 1000
position = 0
positions = [position]

for i in range(steps):
    step = 1 if random.random() > 0.5 else -1
    position += step
    positions.append(position)

plt.figure(figsize=(10, 6))
plt.plot(range(steps+1), positions)
plt.title('1D Random Walk')
plt.xlabel('Step')
plt.ylabel('Position')
plt.grid(True)
plt.show()
```

## 2D Random Walk

In a 2D random walk, the particle can move in four directions: up, down, left, or right. This creates an interesting pattern when visualized.

```python
# 2D Random Walk
steps = 10000
x, y = 0, 0
x_positions, y_positions = [x], [y]

for i in range(steps):
    direction = random.choice(['up', 'down', 'left', 'right'])
    if direction == 'up':
        y += 1
    elif direction == 'down':
        y -= 1
    elif direction == 'left':
        x -= 1
    else:
        x += 1
    
    x_positions.append(x)
    y_positions.append(y)

plt.figure(figsize=(10, 10))
plt.plot(x_positions, y_positions, 'b-', alpha=0.5)
plt.plot(x_positions[0], y_positions[0], 'go', markersize=10, label='Start')
plt.plot(x_positions[-1], y_positions[-1], 'ro', markersize=10, label='End')
plt.title('2D Random Walk')
plt.xlabel('X Position')
plt.ylabel('Y Position')
plt.grid(True)
plt.legend()
plt.axis('equal')
plt.show()
```
